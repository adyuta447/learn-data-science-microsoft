```python
import numpy as np
import matplotlib.pyplot as plt

heights_in_inches = [63, 64, 66, 67, 68, 69, 71, 72, 73, 55, 75]
fig, axs = plt.subplots(1, 2, figsize=(10,5))

axs1 = axs[0]
axs2 = axs[1]

axs1.boxplot(heights_in_inches, whis=0.5)
axs1.set_title('Box plot of heights')
bins = range(min(heights_in_inches), max(heights_in_inches) + 5, 5)

axs2.hist(heights_in_inches, bins=bins, alpha=0.5)
axs2.set_title('Frequency distribution of heights')
axs2.set_xlabel('Height (in)')
axs2.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
