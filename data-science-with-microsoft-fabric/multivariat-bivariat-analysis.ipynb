```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
house_sizes = np.random.randint(1000, 3000, size=50)
house_prices = house_sizes * 100 + np.random.normal(0, 2000, size=50)

plt.scatter(house_sizes, house_prices)
plt.title('House Prices vs Size')
plt.xlabel('Size')
plt.ylabel('Price ($)')

plt.show()
```
